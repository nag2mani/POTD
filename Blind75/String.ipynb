```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        result = 0
        left = 0
        charSet = set()
        for right in range(len(s)):

            while s[right] in charSet:
                charSet.remove(s[left])
                left +=1
                
            charSet.add(s[right])
            result = max(result, right - left + 1)
        
        return result
```

```python
class Solution:
    def characterReplacement(self, s: str, k: int) -> int:
        left = 0
        maxLength = 0
        maxRepeatCount = 0
        map = {}
        for right in range(len(s)):
            map[s[right]] = map.get(s[right], 0) + 1
            maxRepeatCount = max(maxRepeatCount, map.get(s[right], 0))
            windowSize = right - left + 1
            nonRepeatCount = windowSize - maxRepeatCount

            if nonRepeatCount > k:
                map[s[left]] = map.get(s[left], 0) - 1
                left += 1

            maxLength = max(maxLength, right - left + 1)
        # maxLength = max(maxLength, right - left + 1) both are working
        return maxLength
```

```python
from collections import Counter, defaultdict

class Solution:
    def isValid(self, window: dict, t_freq: dict) -> bool:
        """
        Helper function to check if the current window satisfies the condition.
        """
        for char in t_freq:
            if window[char] < t_freq[char]:
                return False
        return True

    def minWindow(self, s: str, t: str) -> str:
        """
        Function to find the minimum window substring.
        """
        if not t or not s:
            return ""

        # Count frequency of characters in t
        t_freq = Counter(t)
        window = defaultdict(int)

        left = 0
        min_len = float('inf')
        min_window = ""

        # Expand the window with the right pointer
        for right in range(len(s)):
            window[s[right]] += 1

            # Check if the window is valid
            while self.isValid(window, t_freq):
                # Update the minimum window if the current one is smaller
                if right - left + 1 < min_len:
                    min_len = right - left + 1
                    min_window = s[left:right + 1]

                # Shrink the window from the left
                window[s[left]] -= 1
                left += 1

        return min_window
```

```python
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False
        a1=Counter(s)
        a2=Counter(t)
        for k in a1:
            if a1[k] != a2[k]:
                return False
        return True
```

```python
from typing import List

class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        hmap={}
        for s in strs:
            k = tuple(sorted(s))
            if k not in hmap:
                a = []
                a.append(s)
                hmap[k] = a
            else:
                hmap[k].append(s)

        return list(hmap.values())
```

```python
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        d = {')' : '(', '}' : '{', ']' : '['}
        for p in s:
            if p in d:
                opening = stack.pop() if stack else "#"
                if d[p] != opening:
                    return False
            else:
                stack.append(p)

        return not stack
```

```python
class Solution:
    def longestPalindrome(self, s: str) -> str:

        def expand_around_center(left: int, right: int) -> str:
            while left >= 0 and right < len(s) and s[left] == s[right]:
                left -= 1
                right += 1
            return s[left + 1 : right]

        longest = ""

        for i in range(len(s)):
            # Odd-length palindromes (single character center)
            odd_palindrome = expand_around_center(i, i)
            if len(odd_palindrome) > len(longest):
                longest = odd_palindrome
                
            # Even-length palindromes (two character center)
            even_palindrome = expand_around_center(i, i + 1)
            if len(even_palindrome) > len(longest):
                longest = even_palindrome

        return longest
```

```python
class Solution:
    def countSubstrings(self, s: str) -> int:

        def expand_around_center(left: int, right: int) -> str:
            count = 0
            while left >= 0 and right < len(s) and s[left] == s[right]:
                count += 1  # Found a palindrome
                left -= 1
                right += 1
            return count

        count = 0
        for i in range(len(s)):
            # Odd-length palindromes (single character center)
            count += expand_around_center(i, i)
            # Even-length palindromes (two character center)
            count += expand_around_center(i, i + 1)

        return count
```
